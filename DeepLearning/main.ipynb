```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.dpi'] = 240
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def hard_sigmoid(x):
    return np.maximum(np.minimum(0.25*x+0.5, np.ones_like(x)), np.zeros_like(x))

def hard_tanh(x):
    return np.maximum(np.minimum(x, np.ones_like(x)), -np.ones_like(x))
```

```python
x = np.linspace(-10, 10, 1000)  # Sample data.

# plt.figure(figsize=(5, 2.7), layout='constrained') #
fig, ax = plt.subplots()
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.

for func in [sigmoid, tanh, hard_sigmoid, hard_tanh]:
    ax.plot(x, func(x), label=str(func.__name__))  # Plot some data on the axes.

ax.legend()  # Add a legend.


```

```python
def ReLU(x):
    return np.maximum(x, np.zeros_like(x))

def Leaky_ReLU(x, gamma = 0.5):
    return np.maximum(x, gamma*x)

def ELU(x, gamma = 0.5):
    return np.maximum(x, np.zeros_like(x))+np.minimum(gamma*(np.exp(x)-1), np.zeros_like(x))

def Softplus(x):
    return np.log(1+np.exp(x))
```

```python
x = np.linspace(-10, 10, 1000)  # Sample data.

# plt.figure(figsize=(5, 2.7), layout='constrained') #
fig, ax = plt.subplots()
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.

for func in [ReLU, Leaky_ReLU, ELU, Softplus]:
    ax.plot(x, func(x), label=str(func.__name__))  # Plot some data on the axes.

ax.legend()  # Add a legend.
```
